```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('ks-projects-201801.csv',low_memory=False, encoding='latin1')
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['usd pledged']=pd.to_numeric(df['usd pledged'],errors='coerce')
```

```python
df['usd pledged']=df['usd pledged'].fillna(df['usd pledged'].mean())
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.head()
```

```python
df=df.drop(columns=['deadline','launched','ID'],axis=1)
```
