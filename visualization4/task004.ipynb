```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/database.csv')
df
```

```python
df.isnull().sum()
```

```python
# Depth Error	18951
# Depth Seismic Stations	16315
# Magnitude Type	3
# Magnitude Error	23085
# Magnitude Seismic Stations	20848
# Azimuthal Gap	16113
# Horizontal Distance	21808
# Horizontal Error	22256
# Root Mean Square	6060
df = df.drop(columns=[
    "Magnitude Error",
    "Horizontal Distance",
    "Horizontal Error",
    "Magnitude Seismic Stations",
    "Depth Error"
])

df
```

```python
df=df.drop(columns=['Depth Seismic Stations', 'Azimuthal Gap'])
df
```

```python
df['Root Mean Square']=df['Root Mean Square'].fillna(df['Root Mean Square'].median())
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Magnitude'].describe()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Magnitude'], bins=30, kde=True)
plt.title('Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Magnitude'])
plt.title('Boxplot of Magnitude')
plt.show()
```

```python
df.head(3)
```

```python
df['Mag_Rounded'] = df['Magnitude'].round()
plt.figure(figsize=(8,5))
sns.countplot(x='Mag_Rounded', data=df)
plt.title('Rounded Magnitude Counts')
plt.show()
```

```python
df[df['Magnitude'] >= 8].shape[0]
```

```python
df['Magnitude'].skew()
```

```python
bins = [5,6,7,8,9,10]
labels = ['5-6','6-7','7-8','8-9','9+']

df['Mag_Category'] = pd.cut(df['Magnitude'], bins=bins, labels=labels, right=False)

sns.countplot(x='Mag_Category', data=df)
plt.title('Magnitude Category Distribution')
plt.show()
```

```python
df.head()
```

```python
iso_rows = df[df["Date"].str.match(r"\d{4}-\d{2}-\d{2}T", na=False)]
```

```python
iso_rows
```

```python
df = df.drop(iso_rows.index)
```

```python
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
numeric_cols = ["Latitude", "Longitude", "Depth", "Magnitude", "Year"]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
```

```python
quakes_per_year = df["Year"].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.plot(quakes_per_year.index, quakes_per_year.values)
plt.title("Earthquakes Per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Depth"], df["Magnitude"], alpha=0.4)
plt.xlabel("Depth")
plt.ylabel("Magnitude")
plt.title("Depth vs Magnitude")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[["Depth","Magnitude","Latitude","Longitude"]].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="Mag_Category", y="Magnitude", data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df["Longitude"], df["Latitude"], alpha=0.3)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Earthquake Locations")
plt.show()
```

```python
avg_mag_year = df.groupby("Year")["Magnitude"].mean()

plt.figure(figsize=(12,6))
plt.plot(avg_mag_year.index, avg_mag_year.values)
plt.title("Average Magnitude Per Year")
plt.show()
```
